```python
from oss import expected_moves
options_file_path = '../../data/options/options.db'
price_file_path = '../../data/prices/stocks.db'
occ = expected_moves(option_db= options_file_path, price_db=price_file_path, verbose = False)
```

```python

from tqdm import tqdm 
import json 

st = '../../data/stocks/tickers.json'
stocks = json.load(open(st))['all_stocks']

lodf = []
contract_list = {}
error = []
for x in tqdm(stocks):
    try: 
        df = occ.run(x, lim = 5)
        df = df[df.expiry == df.expiry.min()]
        lodf.append(df)
    except:
        error.append(x)
        continue
    
allem = pd.concat(lodf)
```

```python
a = allem.set_index(['expiry','expiry1','stock', 'stk_price', 'em', '%','sd','type','strike'])
nxt = a[a.index.get_level_values(1) == '2024-03-15']
#nxt = nxt[(nxt.ask > 0.10) & (nxt.BE > -0.05) & (nxt.BE < 0.05)]
#nxt = nxt[(nxt.index.get_level_values(6) == '2ùùà')]
nxt.sort_values(['ask'], ascending = [False]).head(20).sort_index()
```
