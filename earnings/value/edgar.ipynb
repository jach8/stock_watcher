```python
import edgar
from edgar import Company 
cik, stock = '0001477449','TELADOC HEALTH, INC.'

c = Company(stock, cik)
tree = c.get_all_filings(filing_type = "10-K")
```

```python
docs = c.get_documents(tree, no_of_documents=5)
```

```python
from edgar import TXTML

text = TXTML.parse_full_10K(docs[0])
```
