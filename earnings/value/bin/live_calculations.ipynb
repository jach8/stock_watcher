```python
### Company Valuation Walk Through ####
# Use yfiannce to get company statements 

import json 
import numpy as np 
import pandas as pd 
import datetime as dt 
import yfinance as yf 
import sqlite3 as sql
from tqdm import tqdm 
```

```python
stock = 'gme'
tick = yf.Ticker(stock)
bs = tick.quarterly_balance_sheet
income = tick.quarterly_income_stmt
cf = tick.quarterly_cash_flow
ed = tick.earnings_dates

names = ['balance_sheet', "income_statement", "cashflow", "earnings"]
lodfs = [bs.T, income.T, cf.T, ed]
# lodfs = [df.reset_index() for df in lodfs]

##### Get the data by storing each of the above dataframes in a dictionary with the stock as the key #####
d = {stock: {name: df for name, df in zip(names, lodfs)}}
```

```python
import pickle

# save as pickle
with open(f'{stock}.pkl', 'wb') as f:
    pickle.dump(d, f)
```
