```python
import json 
import pandas as pd 
import numpy as np 
import sqlite3 as sql 

stock_db = sql.connect('../../data/stocks/stock_names.db')
stock_info = json.load(open('../stock_info_all.json', 'r'))
stocks = list(stock_info.keys())
len(stocks)
```

```python
lodf = []; error = []
for x in stocks: 
    try:
        dm = pd.DataFrame(stock_info[x])
        if 'companyOfficers' in dm.columns: 
            dm.drop(columns = ['companyOfficers']) 
        lodf.append(dm)
    except:
        error.append(x)
        
lodf2 = [pd.DataFrame(x.iloc[0]).T for x in lodf]
df = pd.concat(lodf2).reset_index(drop=True)
df.columns = [x.lower() for x in df.columns]

df.beta = df.beta.astype(float) 
df.forwardpe = df.forwardpe.astype(float)
df.columns
```

```python
lodf_etf = []; error2 = []

for i in error:
    try:
        dm = pd.DataFrame(stock_info[i], index = [i])
        lodf_etf.append(dm)
    except:
        error2.append(i)
        
lodf_etf2 = [pd.DataFrame(x.iloc[0]).T for x in lodf_etf]
df_etf = pd.concat(lodf_etf2).reset_index(drop=False).rename(columns = {'index':'stock'})
df_etf.columns
```
